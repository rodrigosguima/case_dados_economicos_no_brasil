```python
# LIBS:

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Lendo a base de dados:

Base_dados = pd.read_excel('Dados_Pib.xlsx')

# Verificando:

Base_dados.head(10)
```

```python
# Sistema de GRIDS:

# Cor de fundo:

Cor_fundo = '#F5F5F5'

# Criar sistema de grids:

Grid_graficos = sns.FacetGrid(Base_dados, col='Territorialidades', hue = 'Territorialidades', col_wrap = 4)

# Adicionar gréfico de linhas em cada gráfico:

Grid_graficos = Grid_graficos.map(plt.plot, 'Ano', 'PIB per capita')

# Adicionar área e ajuste de título nos gráficos:

Grid_graficos = Grid_graficos.map(plt.fill_between, 'Ano', 'PIB per capita').set_titles('{col_name} Territorialidades')

# Filtrar o título:

Grid_graficos = Grid_graficos.set_titles('{col_name}')

# Adicionar subtítulo:

Grid_graficos = Grid_graficos.fig.suptitle('Evolução da renda Per capita por estado', fontsize = 18, color = 'red')

# Ajustando o subtítulo:

plt.subplots_adjust(top = 0.92)
```
